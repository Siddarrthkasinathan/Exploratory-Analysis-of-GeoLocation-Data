Data Exploration and Visualisation

Now that we have our data, we need to understand it. A good way to do this is by visualising the data via graphs. Graphs help us quickly get a sense of the data, and are a much more user friendly way of understanding data as compared to reading thousands of rows of data.

A good graph to look at distributed groups is a "Boxplot". It can tell us at glance where the population is concentrated, and how the outliers compare to the average object in the group.

```python
import numpy as np #NumPy is a widely-used library for numerical and scientific computing in Python. 
from tabulate import tabulate #The tabulate library is used for creating nicely formatted, tabular data from various data structures, such as lists, dictionaries, and more.
import pandas as pd # pandas is a widely used library for data manipulation and analysis.
from sklearn.cluster import KMeans #KMeans class is used for performing k-means clustering, which is an unsupervised machine learning technique for grouping data points into clusters based on their similarity.
import random #The random module is a standard library in Python that provides various functions for generating random numbers and performing randomization-related tasks.
import seaborn as sns #Seaborn is a popular data visualization library that is built on top of Matplotlib and provides a high-level interface for creating informative and attractive statistical graphics. 
import matplotlib.pyplot as plt #Popular Python library for creating static, animated, and interactive visualizations in Python.
import warnings #This code is commonly used when you want to hide warnings that may be generated by various libraries or modules in your code. 
warnings.filterwarnings('ignore')
```

```python
#data cleaning and extracting the relevant features
df1 = pd.read_csv("food_cleaned_data.csv")
df=df1[["cook","diet_current_coded","eating_out","sports","exercise","fav_cuisine_coded","on_off_campus","pay_meal_out","fav_food","fruit_day","income"]]
df.dropna(axis=0,inplace=True)
df.to_csv("food_choices.csv")
df
```

```python
#Plotting Boxplot for the cleaned data(food_choices.csv)
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a boxplot without outliers
sns.boxplot(data=df, palette="Set1", showfliers=False).tick_params(labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.show()
```
